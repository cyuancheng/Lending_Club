## Data Engineering

```python
# Number of each type of column
df.dtypes.value_counts().sort_values()
```

### Categorical variables
Look at how many unique values in each catagorical column

```python
# look at categorical valuables
df[df.select_dtypes('object').columns].head()
```

```python
# see how many unique values in each categorical valuable
df.select_dtypes('object').apply(pd.Series.nunique, axis = 0).sort_values(ascending=False)
```

- remove url, emp_title, zip_code, revol_util,int_rate 
- convert revol_util to num
- convert int_rate to percentage

```python
col_cat_0 =  df.select_dtypes('object').columns.to_list()
col_cat_del = ['url', 'zip_code', 'emp_title', 'revol_util', 'int_rate']
col_cat_0 = list(set(col_cat_0).difference(set(col_cat_del)))
col_cat_0
```

```python
# pick up categorical valuables that are related to default
col_cat = ['term', 'grade', 'emp_length', 'home_ownership',
                    'verification_status', 'loan_status', 'purpose','addr_state',]
```

### Datetime columns
In catagorical columns, three columns are datetime columns 

```python
col_date_0 = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']
```

```python
df[col_date_0].info()
```

```python
df[col_date_0].head()
```

```python
col_date_0
```

```python
# drop missing values
df.dropna(subset=col_date_0, inplace=True)
```

```python
df[col_date_0].isnull().sum()
```

### Percentage columns

```python
col_perc_0 = ['int_rate', 'revol_util']
```

```python
# find missing value
df[col_perc_0].isnull().sum()
```

```python
# drop nan from revol_util column
df.dropna(subset=col_perc_0, inplace=True)
```

```python
df[col_perc_0].isnull().sum()
```

```python
for col in col_perc_0:
    df[col] = df[col].apply(lambda x: x.strip('%'))  # remove %
    df[col] = df[col].apply(lambda x: pd.to_numeric(x))
```

```python
df[col_perc_0].info()
```

### Numerical columns

```python
col_float_0 = df.select_dtypes('float64').columns.to_list()
```

```python
# pick up numerical values related to default
col_float = ['loan_amnt', 'funded_amnt', 'installment', 'annual_inc','int_rate',
                     'dti', 'revol_bal', 'delinq_2yrs', 'open_acc', 'pub_rec',
                                'fico_range_high', 'fico_range_low', 'total_pymnt', 'recoveries']
```
