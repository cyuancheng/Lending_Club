### 1) Deal with datetime columns
In catagorical columns, three columns are datetime columns 

```python
col_date = ['issue_d', 'earliest_cr_line', 'last_pymnt_d']
```

```python
df[col_date].info()
```

```python
# check missing value
df[col_date].isnull().sum()
```

```python
# drop nan 
df.dropna(subset = col_date , inplace=True)
```

```python
df[col_date].isnull().sum()
```

```python
# convert to correct datetime formate
for col in col_date:
    df[col]=df[col].apply(lambda x: datetime.strptime(x,'%b-%Y'))
```

```python
df[col_date].head()
```

```python
df['issue_year'] = df['issue_d'].apply(lambda x: x.year)
```

```python
col_date.append('issue_year')
```

```python
col_date
```

### 2)  Percentage columns

```python
col_perc = ['int_rate', 'revol_util']
```

```python
# find missing value
df[col_perc].isnull().sum()
```

```python
# drop nan from revol_util column
df.dropna(subset=['revol_util'], inplace=True)
```

```python
df[col_perc].isnull().sum()
```

```python
for col in col_perc:
    df[col] = df[col].apply(lambda x: x.strip('%'))  # remove %
    df[col] = df[col].apply(lambda x: pd.to_numeric(x))
```
